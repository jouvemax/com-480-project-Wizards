```python
global_emissions_filtered = global_emissions.drop(columns = ['Data source', 'Gas', 'Unit'])
```

```python
country_code = {}

for country in pycountry.countries:
    country_code[country.name] = country.alpha_3

country_code["Bolivia"] = "BOL"
country_code["Brunei"] = "BRN"
country_code["Cape Verde"] = "CPV"
country_code["Czech Republic"] = "CZE"
country_code["Democratic Republic of the Congo"] = "COD"
country_code["European Union (27)"] = "EU"


country_code["Iran"] = "IRN"
country_code["Laos"] = "LAO"
country_code["Macedonia"] = "MKD"
country_code["Moldova"] = "MDA"
country_code["Micronesia"] = "FSM"
country_code["North Korea"] = "PRK"
country_code["Republic of Congo"] = "COG"
country_code["Russia"] = "RUS"
country_code["South Korea"] = "KOR"
country_code["Syria"] = "SYR"
country_code["Tanzania"] = "TZA"
country_code["Syria"] = "SYR"
country_code["Venezuela"] = "VEN"
country_code["Vietnam"] = "VNM"
country_code["World"] = "WOR"
country_code["South Sudan"] = "SDS"
country_code["Somaliland"] = "ABV"
```

```python
data = {}
from collections import defaultdict

for index, row in global_emissions_filtered.iterrows():
    
    if row['Country'] not in data:
        data[row['Country']] = defaultdict(lambda:0)
    
    country_data = data[row['Country']]
    
    for year in ["2017", "2016", "2015", "2014", "2013", "2012", "2011", "2010", "2009", "2008", "2007", "2006", "2005", "2004", "2003", "2002", "2001", "2000", "1999", "1998", "1997", "1996", "1995", "1994", "1993", "1992", "1991", "1990"]:
        if year not in country_data:
            country_data[year] = defaultdict(lambda: 0)
            
        year_data = country_data[year]
        
        year_data[row["Sector"]] = row[year]
    
clean_dataset = {}
    
for country_name, country_data in data.items():
    
    years_dict = {}
    
    for year, year_data in country_data.items():
        
        if math.isnan(year_data['Electricity/Heat']) and math.isnan(year_data['Building']):
            energy = "NaN"
        else:
            energy = 0
            if not math.isnan(year_data['Electricity/Heat']):
                energy += year_data['Electricity/Heat'] 
            if not math.isnan(year_data['Building']):
                energy += year_data['Building'] 
          
        if math.isnan(year_data['Industrial Processes']) and math.isnan(year_data['Manufacturing/Construction']) and math.isnan(year_data['Fugitive Emissions']): 
            industry = "NaN"
        else:
            industry = 0
            if not math.isnan(year_data['Industrial Processes']):
                industry += year_data['Industrial Processes'] 
            if not math.isnan(year_data['Manufacturing/Construction']):
                industry += year_data['Manufacturing/Construction']
            if not math.isnan(year_data['Fugitive Emissions']):
                industry += year_data['Fugitive Emissions']
         
        if math.isnan(year_data['Agriculture']) and math.isnan(year_data['Land-Use Change and Forestry']) and math.isnan(year_data['Waste']):
            agriculture = "NaN"
        else:
            agriculture = 0
            if not math.isnan(year_data['Agriculture']):
                agriculture += year_data['Agriculture'] 
            if not math.isnan(year_data['Land-Use Change and Forestry']):
                agriculture += year_data['Land-Use Change and Forestry']
            if not math.isnan(year_data['Waste']):
                agriculture += year_data['Waste']
          
        if math.isnan(year_data['Transportation']) and math.isnan(year_data['Bunker Fuels']):
            transport = "NaN"
        else:
            transport = 0
            if not math.isnan(year_data['Transportation']):
                transport += year_data['Transportation'] 
            if not math.isnan(year_data['Bunker Fuels']):
                transport += year_data['Bunker Fuels']
            
        if math.isnan(year_data['Other Fuel Combustion']):
            other = "NaN"
        else:
            other = 0
            if not math.isnan(year_data['Other Fuel Combustion']):
                other += year_data['Other Fuel Combustion'] 
        
            
        #industry = year_data['Industrial Processes'] + year_data["Manufacturing/Construction"] + year_data["Fugitive Emissions"]
        #agriculture = year_data['Agriculture'] + year_data["Land-Use Change and Forestry"] + year_data["Waste"]
        #transport = year_data['Transportation'] + year_data["Bunker Fuels"]
        #other = year_data['Other Fuel Combustion']
        
        if energy == "NaN" and industry == "NaN" and agriculture == "NaN" and transport == "NaN" and other == "NaN":
            total = "NaN"
        else:
            total = 0
            if energy != "NaN":
                total += energy
            if industry != "NaN":
                total += industry
            if agriculture != "NaN":
                total += agriculture
            if transport != "NaN":
                total += transport
            if other != "NaN":
                total += other
            
        
        years_dict[year] = {"energy": energy,
                            "industry": industry,
                            "agriculture": agriculture,
                            "transport": transport,
                            "other": other,
                            "total":total}
        
    
    clean_dataset[country_code[country_name]] = years_dict
        
        
    
#rows = []
#
#for key, val in data.items():
#    
#    country_data = []
#    country_data.append(key)
#    
#    energy = val['Electricity/Heat'] + val['Building']
#    industry = val['Industrial Processes'] + val['Manufacturing/Construction'] + val['Fugitive Emissions']
#    transport = val['Transportation'] + val['Bunker Fuels']
#    agriculture = val['Agriculture'] + val['Land-Use Change and Forestry'] + val['Waste']
#    other =  val['Other Fuel Combustion']
#    
#    country_data.append(energy)
#    country_data.append(industry)
#    country_data.append(transport)
#    country_data.append(agriculture)
#    country_data.append(other)
#    
#    rows.append(country_data)
```
